```python
sensor_readings = ['CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)']
```

```python
for column in sensor_readings:
    plt.figure(figsize=(8, 2))
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
data['DateTime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'])
```

```python
data
```

```python
data.set_index('DateTime', inplace=True)
```

```python
for column in sensor_readings:
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data[column])
    plt.title(f'Time Series of {column}')
    plt.xlabel('DateTime')
    plt.ylabel(column)
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data[sensor_readings + ['T', 'RH', 'AH']].corr(), annot=True, cmap = 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
for column in sensor_readings:
    plt.figure(figsize=(10, 4))
    sns.boxplot(data[column])
    plt.title(f'Outliers in {column}')
    plt.show()
```
